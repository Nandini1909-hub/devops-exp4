```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data


image = data.camera()
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)


sobelx = cv2.Sobel(blurred_image, cv2.CV_64F, 1, 0, ksize=5)  # x direction
sobely = cv2.Sobel(blurred_image, cv2.CV_64F, 0, 1, ksize=5)  # y direction

# Calculate the magnitude of the gradient
abs_sobelx = np.absolute(sobelx)
abs_sobely = np.absolute(sobely)
grad = np.hypot(abs_sobelx, abs_sobely)

# Convert the gradient to 8-bit for display
grad_8u = np.uint8(grad)

# Results
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(blurred_image, cmap='gray')
plt.title('Blurred Image')

plt.subplot(1, 3, 2)
plt.imshow(abs_sobelx, cmap='gray')
plt.title('Sobel X')

plt.subplot(1, 3, 3)
plt.imshow(abs_sobely, cmap='gray')
plt.title('Sobel Y')

plt.show()

plt.imshow(grad_8u, cmap='gray')
plt.title('Gradient Magnitude')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data


image = data.camera()
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)


sobelx = cv2.Sobel(blurred_image, cv2.CV_64F, 1, 0, ksize=5)  # x direction
sobely = cv2.Sobel(blurred_image, cv2.CV_64F, 0, 1, ksize=5)  # y direction

# Calculate the magnitude of the gradient
abs_sobelx = np.absolute(sobelx)
abs_sobely = np.absolute(sobely)
grad = np.hypot(abs_sobelx, abs_sobely)

# Convert the gradient to 8-bit for display
grad_8u = np.uint8(grad)

# Results
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(blurred_image, cmap='gray')
plt.title('Blurred Image')

plt.subplot(1, 3, 2)
plt.imshow(abs_sobelx, cmap='gray')
plt.title('Sobel X')
plt.subplot(1, 3, 3)
plt.imshow(abs_sobely, cmap='gray')
plt.title('Sobel Y')

plt.show()

plt.imshow(grad_8u, cmap='gray')
plt.title('Gradient Magnitude')
plt.show()
alpha = 1.5  # Contrast control (1.0 is no change)
beta = 50   # Brightness control (0 is no change)
abs_sobelx = np.absolute(sobelx)
adjusted_sobelx = cv2.convertScaleAbs(abs_sobelx, alpha=alpha, beta=beta)

# Adjust brightness and contrast for Sobel Y
alpha = 1.5  # Contrast control (1.0 is no change)
beta = 50   # Brightness control (0 is no change)
abs_sobely = np.absolute(sobely)
adjusted_sobely = cv2.convertScaleAbs(abs_sobely, alpha=alpha, beta=beta)


# Display results
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(adjusted_sobelx, cmap='gray')
plt.title('Adjusted Sobel X')

plt.subplot(1, 2, 2)
plt.imshow(adjusted_sobely, cmap='gray')
plt.title('Adjusted Sobel Y')

plt.show()
#Gradient Magnitude
Gradient_Magnitude = np.sqrt(sobelx**2 + sobely**2)
#display the results
plt.imshow(Gradient_Magnitude, cmap='gray')
plt.title('Gradient Magnitude')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data

# Load the image
image = data.camera()

# Apply Sobel operator with different kernel sizes
sobelx3 = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobely3 = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
sobelx5 = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5)
sobely5 = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)
sobelx7 = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=7)  # Sobel X with kernel size 7
sobely7 = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=7)  # Sobel Y with kernel size 7


# Display the results
plt.figure(figsize=(18, 5))  # Adjust figure size

plt.subplot(1, 7, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 7, 2)
plt.imshow(sobelx3, cmap='gray')
plt.title('Sobel X-3 Image')

plt.subplot(1, 7, 3)
plt.imshow(sobely3, cmap='gray')
plt.title('Sobel Y-3 Image')

plt.subplot(1, 7, 4)
plt.imshow(sobelx5, cmap='gray')
plt.title('Sobel X-5 Image')  # Title for Sobel X-7

plt.subplot(1, 7, 5)
plt.imshow(sobely5, cmap='gray')
plt.title('Sobel Y-5 Image')  # Title for Sobel Y-7

plt.subplot(1, 7, 6)
plt.imshow(sobelx7, cmap='gray')
plt.title('Sobel X-7 Image')

plt.subplot(1, 7, 7)
plt.imshow(sobely7, cmap='gray')
plt.title('Sobel Y-7 Image')

plt.tight_layout()
plt.show()
```

**Analysis:**

Kernel Size 5 (sobelx5, sobely5):

Detects edges relatively well, providing a good balance between edge detection and noise reduction.
Edges appear sharper and more defined compared to the original image.
Might not capture very fine details due to the moderate kernel size.
Kernel Size 7 (sobelx7, sobely7):

Detects thicker edges compared to kernel size 5.
May introduce more noise or artifacts due to the larger kernel size.
Edges are less sharp and may have a slightly blurred appearance.
Kernel Size 11 (sobelx11, sobely11):

Detects even thicker edges and highlights larger structures in the image.
May further increase noise and blur the edges more significantly.
Fine details are lost due to the aggressive smoothing effect of the large kernel.

Conclusion:

The best Sobel filter among the three is likely the one with kernel size 5 (sobelx5, sobely5). It offers a good trade-off between edge detection, sharpness, and noise reduction. It effectively highlights edges without introducing excessive noise or blurring.
